# Machine Learning - Scatter Plot

<i>A scatter plot is a diagram where each value in the data set is represented by a dot.</i>

Here, we will take two arrays and the first one represents the age of cars and the last array represents the speed of those car

```python
import matplotlib.pyplot as mpl

x = [4,5,2,9,5,5,3,1,1,7,2,4,5,6,3,8,2,1,7,4]
print(f"Number of arrays in {len(x)}")

y = [85,75,96,60,86,88,98,112,117,64,91,88,73,65,88,63,69,105,68,93]
print(f"Number of arrays in {len(y)}")

mpl.scatter(x,y)
mpl.show()
```

## Random Data Distribution - Scatter Plot

```python
import numpy as np

x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10, 2.0, 1000)

mpl.scatter(x, y)
mpl.show()
```

```python
import matplotlib.pyplot as plt
import random
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

mycolor = random.sample(range(1, 100),len(x))
plt.scatter(x, y, c = mycolor, cmap = "viridis")
plt.colorbar()
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.normal(7.0, 1.0, 1000)
y = numpy.random.normal(14.0, 2.0, 1000)
mycolor = random.sample(range(-1, 1000),len(x))
plt.scatter(x, y, c = mycolor, cmap = "viridis")
plt.colorbar()
plt.show()
```
