```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

plt.scatter(x, y)
plt.show() 
```

Now we compute the ward linkage using euclidean distance, and visualize it using a dendrogram:

```python
data =list(zip(x, y))
```

```python
data
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
linkagedata = linkage(data, method = "ward", metric = "euclidean")
```

```python
dendrogram(linkagedata)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hierarchicalcluster = AgglomerativeClustering(n_clusters = 2, affinity = "euclidean", linkage = "ward")
```

```python
labels = hierarchicalcluster.fit_predict(data)
```

```python
labels
```

```python
len(labels)
```

```python
plt.scatter(x, y, c = labels)
plt.show()
```
