```python
import pandas as pd
```

```python
df = pd.read_csv("cardata.csv")
```

```python
df.head()
```

```python
X = df[["Volume", "Weight"]]
y = df["CO2"]
```

### Scaling is need to make over all the data has same figure

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scalexy = scale.fit_transform(X, y)
```

### Linear Regression

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(scalexy, y)
```

### Random values for prediction in scale transformation

```python
scaled = scale.transform([[2.26, 1532]])
```

```python
predict_co2 = regr.predict([scaled[0]])
```

```python
print(predict_co2)
```
