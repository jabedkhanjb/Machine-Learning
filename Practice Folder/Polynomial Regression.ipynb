```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = np.array([1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22])
y = np.array([100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100])
```

```python
x
```

```python
y
```

```python
colors = random.sample(range(1, 100), len(x))
```

```python
plt.scatter(x, y, c = colors, cmap="viridis")
plt.colorbar()
plt.grid()
plt.show()
```

#### Linear Line  

```python
mymodel = np.poly1d(np.polyfit(x, y, 3))
myline = np.linspace(1, 22, 100)
plt.plot(myline, mymodel(myline), color = "m")
plt.scatter(x, y, c=colors,cmap="viridis")
plt.show()
```

# R-Squared

```python
from sklearn.metrics import r2_score
```

```python
mymodel
```

```python
print(r2_score(y, mymodel(x)))
```

# Predict Future Values

```python
mymodel
```

```python
speed = mymodel(4)
speed
```

```python
# now let's demonstrate the plot again
plt.scatter(x, y, c=colors, cmap = "viridis")
plt.xlim(1, 20)
plt.plot(myline, mymodel(myline), color="m")
plt.plot(x, y, c="y")
plt.grid()
print(speed)
plt.show()
```
