# Code 1

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
```

```python
x = np.random.normal(7, 8, 100)
y = np.random.normal(110, 60, 100) / x
```

```python
plt.scatter(x, y)
plt.show()
```

### Split into train test 

```python
# x, y train (o to 70 out of 100)
xtrain = x[:70]
ytrain = y[:70]

#x, y test (70 to remaining value which is till 100)
xtest = x[70:]
ytest = y[70:]
```

```python
# display x, y train
plt.scatter(xtrain, ytrain)
plt.show()
```

```python
# display x, y test
plt.scatter(xtest, ytest)
plt.show()
```

### Fit the data set with polynomial regression

```python
mymodel = np.poly1d(np.polyfit(xtrain, ytrain, 2))
myline = np.linspace(0, 10, 100)
```

```python
plt.scatter(xtrain, ytrain)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
relation = r2_score(ytrain, mymodel(xtrain))
```

```python
print(relation)
```

###### So here the relation is very bad. we can't proceed further.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
```

```python
x = np.random.normal(8, 1, 100)
y = np.random.normal(190, 20, 100) / x
```

```python
xtrain = x[:70]
ytrain = y[:70]

xtest = x[70:]
ytest = y[70:]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(xtrain, ytrain)
plt.show()
```

```python
plt.scatter(xtest, ytest)
plt.show()
```

the best way to do use polynomial regression part on this code

```python
mymodel = np.poly1d(np.polyfit(xtrain, ytrain, 3))
```

```python
myline = np.linspace(5, 10, 100)
```

```python
plt.scatter(xtrain, ytrain)
plt.plot(myline, mymodel(myline))
```

```python
from sklearn.metrics import r2_score
```

### Relationship Score in Training

```python
trainrelation = r2_score(ytrain, mymodel(xtrain))
```

```python
trainrelation
```

### Relationship Score in Testing

```python
testrelation = r2_score(ytest, mymodel(xtest))
```

```python
testrelation
```

both have a good relationship, so now we can confidently use the prediction on this model
