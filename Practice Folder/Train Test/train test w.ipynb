# Code 1

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
```

```python
x = np.random.normal(5, 2, 100)
y = np.random.normal(180, 60, 100)
```

```python
plt.scatter(x, y)
plt.show()
```

### Split into train test 

```python
# x, y train (o to 70 out of 100)
xtrain = x[:70]
ytrain = y[:70]

#x, y test (70 to remaining value which is till 100)
xtest = x[70:]
ytest = y[70:]
```

```python
# display x, y train
plt.scatter(xtrain, ytrain)
plt.show()
```

```python
# display x, y test
plt.scatter(xtest, ytest)
plt.show()
```

### Fit the data set with polynomial regression

```python
mymodel = np.poly1d(np.polyfit(x, y, 5))
myline = np.linspace(0, 10, 100)
```

```python
plt.scatter(xtrain, ytrain)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
relation = r2_score(ytrain, mymodel(xtrain))
```

```python
print(relation)
```

###### So here the relation is very bad. we can't proceed further.
