```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
```

```python
x = np.random.normal(5, 2, 100)
y = np.random.normal(110, 50, 100) / x
```

```python
plt.scatter(x, y)
plt.show()
```

# Split Train Test

```python
xtrain = x[:80]
ytrain = y[:80]

xtest = x[80:]
ytest = y[80:]
```

```python
plt.scatter(xtrain, ytrain)
plt.show()
```

### Try out with Polynomial Regression

```python
mymodel = np.poly1d(np.polyfit(xtrain, ytrain, 3))
```

```python
myline = np.linspace(0, 10, 100)
```

```python
plt.scatter(xtrain, ytrain)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
relationship = r2_score(ytrain, mymodel(xtrain))
```

```python
relationship
```

```python
print(f"Since the relationship '{relationship}' is not between 0 and 1 even not near about 1. \nSo it is consided as a bad relationship")
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2)
```

```python
x = np.random.normal(4, 1, 100)
y = np.random.normal(140, 45, 100) / x
```

```python
plt.scatter(x, y)
plt.show()
```

```python
xtrain = x[:70]
ytrain = y[:70]

xtest = x[70:]
ytest = y[70:]
```

```python
mymodel = np.poly1d(np.polyfit(xtrain, ytrain, 3))
myline = np.linspace(0, 7, 100)
```

```python
plt.scatter(xtrain, ytrain)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
relationship = r2_score(ytrain, mymodel(xtrain))
```

```python
relationship
```

```python
relationship2 =r2_score(ytest, mymodel(xtest))
relationship2
```

Train and Test both have a good relationship
