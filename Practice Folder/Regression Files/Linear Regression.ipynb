```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
colors = random.sample(range(1, 100), len(x))
```

```python
from scipy import stats
```

```python
slope, intercept, r, p, str_err = stats.linregress(x, y)
```

```python
def myfunction(jb):
    return slope * jb + intercept

mymodel = list(map(myfunction, x))
```

```python
plt.scatter(x, y, c = colors, cmap="viridis")
plt.plot(x, mymodel)
plt.show()
```

```python
print(r)
```

```python
speed = myfunction(9)
print(speed)
```
