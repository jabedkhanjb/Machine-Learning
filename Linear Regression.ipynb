### Linear Regression

##### Firstly we will find the scatter plot by using the data of n number of car's age and their speed level

```python
import matplotlib.pyplot as mpl

x = [2, 3, 5, 2, 4, 6, 11, 1, 16, 22, 12, 19]
y = [102, 105, 89, 99, 92, 82, 67, 109, 60, 51, 72, 59]

mpl.scatter(x, y)
mpl.show()
```

```python
import matplotlib.pyplot as mpl
from scipy import stats

x = [2, 3, 5, 2, 4, 6, 11, 1, 16, 22, 12, 19]
y = [102, 105, 89, 99, 92, 82, 67, 109, 60, 51, 72, 59]
# x = [67, 53, 32, 40, 35]
# y = [95, 97, 80, 85, 91]

slope, intercept, r, p, std_err = stats.linregress(x, y)
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"r for Relationship: {r}")
print(f"P: {p}")
print(f"STD_ERR: {std_err}")

def myfunction(x):
    return slope * x + intercept

mymodel = list(map(myfunction, x))

mpl.scatter(x, y)
mpl.plot(x, mymodel)
mpl.show()
```

```python
import matplotlib.pyplot as plt
import random
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

mycolor = random.sample(range(1,100),len(x))

# linear regression
# need to import stats from scipy

slope, intercept, r, p, std_err = stats.linregress(x,y)

def myfunction(x):
    return slope * x + intercept

mymodel = list(map(myfunction, x))


plt.scatter(x,y,c=mycolor, cmap="viridis")
plt.colorbar()
plt.plot(x,mymodel)
plt.show()
```
